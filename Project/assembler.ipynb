```python
read_file = open("cse141l.txt", "r")
write_file = open("machine_code.txt","w")
f = read_file.readlines()
```

```python
op_map = {}
op_map["MOV"] = "000"
op_map["SLL"] = "000"
op_map["SRL"] = "000"
op_map["HALT"] = "000"
op_map["AND"] = "001"
op_map["XOR"] = "010"
op_map["LD"] = "011"
op_map["SW"] = "100"
op_map["BNE"] = "101"
op_map["ADD"] = "110"
op_map["JR"] = "111"
op_map["INC"] = "111"

jr_table = {}
jr_table["LOOP1_1"] = "00000"
jr_table["LOOP2_1"] = "00001"
jr_table["LOOP2_2"] = "00010"
jr_table["ELSE2_1"] = "00011"
jr_table["ELSE2_2"] = "00100"
jr_table["LOOP2_3"] = "00101"
jr_table["LOOP3_1_1"] = "00110"
jr_table["ELSE3_1_1"] = "00111"
jr_table["ELSE3_1_2"] = "01000"
jr_table["ELSE3_1_3"] = "01001"
jr_table["ELSE3_1_4"] = "01010"
jr_table["ELSE3_1_5"] = "01011"
jr_table["LOOP3_2_1"] = "01100"
jr_table["ELSE3_2_1"] = "01101"
jr_table["ELSE3_2_2"] = "01110"
jr_table["ELSE3_2_3"] = "01111"
jr_table["ELSE3_2_4"] = "10000"
jr_table["ELSE3_2_5"] = "10001"
jr_table["END3_2"] = "10010"
jr_table["LOOP3_3_1"] = "10011"
jr_table["ELSE3_3_1"] = "10100"
jr_table["ELSE3_3_2"] = "10101"
jr_table["ELSE3_3_3"] = "10110"
jr_table["ELSE2_4"] = "11110"
jr_table["ELSE2_5"] = "11111"

line_count = 0
print_line = False

for line in f:
    try:
        key = line.split()[0]
    except:
        continue
    
    if key[0:2] == "//":
        continue
        
    if key in op_map:
        op = op_map[key]
        
        if key == "MOV":
            rs = line.split()[1]
            imm = line.split()[2]
            rs_code = format(int(rs[1]), '03b')
            imm_code = format(int(imm), '01b')
            func_code = format(0, '02b')
            machine_code = op + rs_code + imm_code + func_code
            if print_line == True:
                print(key, rs, imm, "line: ", line_count)
                print_line = False
#             print(key, rs, imm)
#             print(machine_code)
        
        if key == "SLL":
            rs = line.split()[1]
            imm = line.split()[2]
            rs_code = format(int(rs[1]), '03b')
            imm_code = format(int(imm), '01b')
            func_code = format(1, '02b')
            machine_code = op + rs_code + imm_code + func_code
            if print_line == True:
                print(key, rs, imm, "line: ", line_count)
                print_line = False
#             print(key, rs, imm)
#             print(machine_code)
            
        if key == "SRL":
            rs = line.split()[1]
            imm = line.split()[2]
            rs_code = format(int(rs[1]), '03b')
            imm_code = format(int(imm), '01b')
            func_code = format(2, '02b')
            machine_code = op + rs_code + imm_code + func_code
            if print_line == True:
                print(key, rs, imm, "line: ", line_count)
                print_line = False
#             print(key, rs, imm)
#             print(machine_code)
        
        if key == "HALT":
            rs_code = format(0, '03b')
            imm_code = format(0, '01b')
            func_code = format(3, '02b')
            machine_code = op + rs_code + imm_code + func_code
            if print_line == True:
                print(key, "line: ", line_count)
                print_line = False
#             print(key)
#             print(machine_code)
        
        if (key == "AND" or key == "XOR" or key == "LD" or key == "SW" or key == "BNE" or key == "ADD") :
            rs = line.split()[1]
            rt = line.split()[2]
            rs_code = format(int(rs[1]), '03b')
            rt_code = format(int(rt[1]), '03b')
            machine_code = op + rs_code + rt_code
            if print_line == True:
                print(key, rs, rt, "line: ", line_count)
                print_line = False
#             print(key, rs, rt)
#             print(machine_code)
        
        if key == "JR":
            address = jr_table[line.split()[1]]
            func_code = format(0, '01b')
            machine_code = op + address + func_code
            if print_line == True:
                print(key, line.split()[1], "line: ", line_count)
                print_line = False
#             print(key, line.split()[1])
#             print(machine_code)   
        
        if key == "INC":
            rs = line.split()[1]
            rs_code = format(int(rs[1]), '05b')
            func_code = format(1, '01b')
            machine_code = op + rs_code + func_code
            if print_line == True:
                print(key, rs, "line: ", line_count)
                print_line = False
#             print(key, rs)
#             print(machine_code)
        
        line_count = line_count + 1
        write_file.write(machine_code)
        write_file.write("\n")    
    
    else:
        print_line = True
        print(key)
```
